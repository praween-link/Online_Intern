## Hypothesis Testing in Month

**Filter Month**

```python
# Filter Months
def filter_month(ndata, d1, d2):
    tag = ndata[ndata['Formatted Date'] >= f'{d1}-01 01:00:00.000 +0200']
    tg = tag[tag['Formatted Date'] < f'{d2}-01 01:00:00.000 +0200']
    return tg
```

```python
# one_month_plot = one_month.drop(['Wind Bearing (degrees)', 'Temperature (C)', 'Visibility (km)', 'Loud Cover', 'Pressure (millibars)', 'Daily Summary', 'Wind Speed (km/h)', 'Summary', 'Precip Type'], axis=1)
# one_month_plot.columns
ndata.head()
```

### 1.) Graph: 'Apparent Temperature (C)' and 'Humidity' of 'January 2006'

**Dataset of 'January 2006'**

```python
one_2006_Jun = filter_month(ndata, '2006-01', '2006-02')
one_2006_Jun.shape
```

**Plot 'Apparent Temperature (C)' and 'Humidity'**

```python
one_2006_Jun.set_index('Formatted Date')['Apparent Temperature (C)'].plot(figsize=(15, 5), linewidth=1, color='maroon')
one_2006_Jun.set_index('Formatted Date')['Humidity'].plot(figsize=(15, 5), linewidth=1, color='blue')
plt.legend()
plt.title("Apparent Temperature (C) and Humidity of 1-month January-2006")
plt.ylabel("Apparent Temperature (C) and Humidity")
```

**Hypothesis**

```python
def hypothesis(weather_data):
    stat, p, lags, obs, crit, t = adfuller(weather_data)

    stat_p = 'stat=%.3f, p=%.3f' % (stat, p)
    if p > 0.05:
        hyth = "Null Hypothesis"
    else:
        hyth = "Alternative Hypothesis"
    
    return stat_p, hyth
```

```python
hypothesis_r = hypothesis(one_2006_Jun['Apparent Temperature (C)'])
print(hypothesis_r[0])
print(f'Hypothesis result: {hypothesis_r[1]}')
```

**Average Apparent temperature and humidity**

```python
t_avg = one_2006_Jun['Apparent Temperature (C)'].mean()
h_avg = one_2006_Jun['Humidity'].mean()
print(f" Average of Apparent temperature: {t_avg} \n Average of Humidity: {h_avg}")
```

It also provides a compelling new motivation for forest conservation. So, that hypothesis provide two results 'Null Hypothesis' and 'Alternative Hypothesis'. we may easuly know weather.<br>
And we have known from this graph, in january 2006, '**Apparent Temperature**' less than '**Humidity**'. So, we can say that in this month rained approximately 85% - 90%. Where 'Formatted Date' on x-axis and 'Apparent Temprature' and 'Humidity' on y-axis.
<br>
We are caluculate result of hypothesis is *Alternative Hypothesis* where, stat = -3.595 and  p = 0.006.<br>
Avarage of Appratent temprature is -4.175164277180406 and avarage of Humidity is 0.8346639784946237.
